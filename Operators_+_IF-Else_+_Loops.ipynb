```python
# Arithmetic Operators
print(5+6)

print(5-6)

print(5*6)

print(5/6)

print(5//2)

print(5%2)

print(5**2)
```

```python
# Relational Operators
print(4>5)

print(5<2)

print(5>=2)

print(5<=2)

print(5==2)

print(5!=2)
```

```python
# Logical Operators
print(0 and 1)

print(1 or 0)

print(not 0)
```

```python
# Bitwise Operators
# 2 -> 010
# 3 -> 110


# bitwise and operator
# 010 - > 2
print(2 & 3)

# bitwise or operator
# 110 -> 3
print(2 | 3)

# bitwisw X0R operator
# 100 -> 1
print(2 ^ 1)

# bitwise Not
print(~3)

# leftshift
print(4 >> 2)

# rightshift
print(5 << 2)
```

```python
# Assignment Operator
# '=' 
a = 2

#a = a+2
a += 2

print(a)
```

```python
# Membership Operators

# in / not in

print('D' in 'Delhi')
print('D' not in 'Delhi')

print(1 in [2,3,4])
```

```python
# Guessing Game

# Generate a random integer between 1 and 100
import random
jackpot = random.randint(1,100)

counter = 1

try:
    guess = int(input('Guess Karo: '))

    while guess != jackpot:
        if guess < jackpot:
            print('Thoda uper guess karo')
        elif guess > jackpot:
            print('Thoda neeche guess karo')
        guess = int(input('Galat tha! Firse: '))
        counter += 1
    else:
        print('Sahi pakde ho')
        print(counter, 'me converge kiye tum ')
except:
    print('Ye gawar hai! Number daalo bhaisahab')
```

## Sequence Sum

```python
# 1/1! + 2/2! + 3/3! +...

n = int(input('enter n: '))

result=0
fact = 1

for i in range(1,n+1):
    
    fact = fact*i
    
    result = result + i/fact
    
print(result)
    
```

# Pattern

```python
rows = int(input('Enter num: '))

# for i in range(1,rows+1):
#     print(i*'*')

for i in range(1,rows+1):
    for j in range(1,i+1):
        print('*',end='')
    print()
```

```python
rows = int(input('Enter num: '))

for i in range(1,rows+1):
    for j in range(1,i+1):
        print(j, end='')
    for k in range(i-1,0,-1):
        print(k, end='')
    print()
```

# Loop Control Statement
- Break
- Continue
- Pass

```python
# break
for i in range(1,10):
    if i == 5: 
        break
    print(i)
```

```python
# prime number
l_r = int(input('Enter upper num: '))
u_r = int(input('Enter lower num: '))

for i in range(l_r,u_r+1):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        print(i)
```

```python
# Continue
for i in range(1,10):
    if i == 5:
        continue
    print(i)
```

```python
for i in range(1,10):
    pass
```
